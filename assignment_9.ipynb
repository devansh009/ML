```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/house-price-prediction/master/kc_house_data.csv ")
```

```python
df
```

```python
x=df.iloc[:,13].values
y=df.iloc[:,2].values
```

```python
x
```

```python
y
```

```python
df.plot(y='price',x='yr_built',style='o')
plt.xlabel('sqft_living')
plt.ylabel('price')
plt.title('assignment')
plt.show()
```

```python
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=3/10,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
X_train =x_train.reshape(15129,1)
y_train =y_train.reshape(15129,1)
X_test  =x_test.reshape(6484,1)
y_test  =y_test.reshape(6484,1)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
df1=pd.DataFrame({'Actual:':list(y_test),'Predicted:':list(y_pred)})
df1
```

```python
plt.scatter(x='X_train',y='y_train',color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.show()
```

```python
plt.scatter(x='X_test',y='y_test',color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.show()
```

```python
from sklearn import metrics
print("Mean Absolute error     :",metrics.mean_absolute_error(y_test,y_pred))
print("Mean Squared error      :",metrics.mean_squared_error(y_test,y_pred))
print("Root mean absolute error:",np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```
